```python
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression 
```

```python
data = pd.read_csv('09.1 lsd_math_score_data.csv')
data
```

```python
time = data['Time_Delay_in_Minutes']
lsd = data[['LSD_ppm']]
score = data[['Avg_Math_Test_Score']]
```

```python
%matplotlib inline

plt.plot(time, lsd, color='#e67e22', linewidth=2)
plt.title('Tissue concentration of LSD overtime', fontsize= 16) 
plt.xlabel('Time in minutes', fontsize = 14)
plt.ylabel('tissue LSD ppm', fontsize=14)
plt.xlim(0, 500) #puts lim in x-axis
plt.ylim(1,7)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.text(0, -0.5, 'wagner at el (1968)', fontsize=12, color='#e74c3c')
plt.style.use('fivethirtyeight')
```

```python
regression = LinearRegression().fit(lsd, score)
print(regression.coef_[0][0])
print(regression.intercept_[0])
print(regression.score(lsd, score))
```

```python
plt.scatter(lsd, score, color='blue', alpha=0.7,s=100)
plt.style.use('fivethirtyeight')
plt.title('LSD-25 vs Average Math Test Score', fontsize=16)
plt.xlabel('LSD ppm', fontsize=14)
plt.ylabel('Math Test Score', fontsize=14)
plt.xlim(1,8)
plt.ylim(20,85)
plt.plot(lsd, regression.predict(lsd), color='#e74c3c')
```
