```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv('cost_revenue_cleann.csv')
```

```python
data
```

```python
data.describe()
```

```python
X = DataFrame(data, columns = ['production_budget_usd'])
y = DataFrame(data, columns = ['worldwide_gross_usd'])
```

```python
X.iloc[0:10, :1]
```

```python
plt.figure(figsize = (12, 6))
plt.scatter(X, y, alpha = 0.3)
plt.title('movie revenue vs production budget')
plt.xlabel('production budget $')
plt.ylabel('worldwide gross $')
plt.xlim(0, 450000000)
plt.ylim(0, 3000000000)
plt.show()
```

```python
regression = LinearRegression()
```

finding the best fit of given data

```python
regression.fit(X, y)
```

slope of best fit line:

```python
regression.coef_
```

slope 3.12 means for every 1usd invested we get 3.12 usd in average which is not bad

intercept of best fit line:

```python
regression.intercept_
```

intercept -7million means that 1000 usd budget movie are loosing 7 million which doesn't make sense

```python
regression.score(X, y)
```

```python
plt.figure(figsize = (12, 6))
plt.scatter(X, y, alpha = 0.3)

plt.plot(X, regression.predict(X), color='red', linewidth='4')

plt.title('movie revenue vs production budget')
plt.xlabel('production budget $')
plt.ylabel('worldwide gross $')
plt.xlim(0, 450000000)
plt.ylim(0, 3000000000)
plt.show()
```
